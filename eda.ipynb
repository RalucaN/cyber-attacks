```python
import pandas as pd
import json

events='datasets/events.json'
actors='datasets/actors.json'


def json_to_df(filename):
    with open(filename, 'r') as jread:
        events_json=json.load(jread)
    df=pd.json_normalize(events_json['rows'])
    return df


events_df=json_to_df(events)
actors_df=json_to_df(actors)
```
